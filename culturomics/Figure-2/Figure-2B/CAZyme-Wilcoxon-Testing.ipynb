```r
library(tidyr)
library(ggplot2)
library(ggh4x)
library(dplyr)
library(RColorBrewer)
library(tidyselect)
library(tidyr)
library(stringr)
library(ggpubr)
library(labdsv)
library(vegan)
library(readr)
library(ape)
library(lme4)
library(lsmeans)
library(scales)
library(igraph)
```

```r
comparison_groups = list(c("Cow", "Farmer"),c("Non-Farmer", "Farmer"), c("Cow", "Non-Farmer") )
```

```r
total_cazymes_dereplicated_isolates <- read.csv("DOME-Dereplicated-Isolates-Total-CAZymes.csv")
msa_total_cazymes_dereplicated <- total_cazymes_dereplicated_isolates %>% filter(Media == "MSA")
```

```r
msa_total_cazyme_wilcox_results <- compare_means(
  CAZyme.Count ~ Cohort,
  data = msa_total_cazymes_dereplicated,
  method = "wilcox.test",
  comparisons = comparison_groups,
  p.adjust.method = "BH"
)

msa_total_cazyme_wilcox_results <- msa_total_cazyme_wilcox_results %>%
select(group1, group2, p.adj) %>%
rename(
    Group1 = group1,
    Group2 = group2
)
```

```r
write.csv(msa_total_cazyme_wilcox_results, "DOME-MSA-Total-CAZyme-Cohort-Wilcoxon-Results.csv", row.names = FALSE)
```
